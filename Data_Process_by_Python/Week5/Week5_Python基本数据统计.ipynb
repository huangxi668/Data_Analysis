```python
#完整的程序
from matplotlib.finance import quotes_historical_yahoo_ochl
from datetime import date
from datetime import datetime
import pandas as pd
today = date.today()
start = (today.year-1, today.month, today.day)
quotes = quotes_historical_yahoo_ochl('AXP', start, today)
fields = ['date','open','close','high','low','volume']
list1 = []
for i in range(0,len(quotes)):
    x = date.fromordinal(int(quotes[i][0]))
    #把日期转换成常规格式
    y = datetime.strftime(x,'%Y-%m-%d')
    #转换成固定格式，%Y表示年完整显示，%y只显示后两位
    list1.append(y)
    #增加日期列表的长度
quotesdf = pd.DataFrame(quotes, index = list1, columns = fields)
quotesdf = quotesdf.drop(['date'], axis = 1)
#删除原来date列
print quotesdf
```

### 5.4数据选择
* 选择方式：
 选择行:切片；索引  
 选择列：列名；行、列（标签label（loc））   
 行和列的区域：标签label（loc）；iloc（位置）   
 单个值：at；iat(索引必须是整数)   
 注意loc,iloc,at,iat,ix用法的区别，具体百度pandas 。。。。   
 条件筛选  

```python
print quotesdf[u'2016-02-08 ':u'2016-02-10 ']
#直接根据索引值来索引相关内容，索引方式直接
#u表示Unicode，如果缺少了就无法找到
print quotesdf['open'][1:3]
print quotesdf.close[2:4]
#根据列名选择,会打印出索引值
print quotesdf.loc[u'2016-02-08 ':u'2016-02-10 ',]
#df.loc[rows,columns] 根据行标签和列标签进行选择，row,columns参数是可以选择的
print quotesdf.loc[u'2016-02-08 ':u'2016-02-10 ',['open','high']]
#print quotesdf.iloc[u'2016-02-08 ':u'2016-02-10 ',['open','high']]
#TypeError: cannot perform reduce with flexible type
#选择块区域
#print quotesdf.loc[:,['open','high']]
#  ":"不能缺

#选择单个值
print quotesdf.loc[u'2016-02-08','open']
print quotesdf.at[u'2016-02-08','open']
#print quotesdf.iat[u'2016-02-08','open']
#ValueError: iAt based indexing can only have integer indexers

#条件选择
print quotesdf[quotesdf.index>=u'2017-01-05']
print quotesdf[(quotesdf.index>=u'2016-08-05')&(quotesdf.close>=75)]
```

```python
#求开盘价的平均值 
#print quotesdf.mean(columns = 'open')
#TypeError: mean() got an unexpected keyword argument 'columns'

#收盘价高于开盘价的天数
print len(quotesdf[quotesdf.close > quotesdf.open])
print len(quotesdf)-len(quotesdf[quotesdf.close > quotesdf.open])

#统计相邻两天收盘价的涨跌情况
status = np.sign(np.diff(quotesdf.close))
print status
print status[np.where( status == 1.)].size
#1 表示涨
print status[np.where( status == -1.)].size
# -1表示跌

#排序  sort函数有很多问题？？？
print quotesdf.sort(columns='open')[:10]
#前10天开盘价排序,默认是升序排列
#DataFrame的sort()函数已不推荐使用，相同功能推荐使用sort_index()函数
#re_quotesdf_sort = quotesdf['open'][:10]
#print re_quotesdf_sort.sort(ascending = False)
#可以添加sort()函数的ascending属性控制顺序/逆序排序，默认该属性=True，即顺序排列
#TypeError: sort() got an unexpected keyword argument 'reverse'
#print quotesdf.sort(columns='open')[:10].close
```

```python
#计数统计
#统计2016年2月份开盘天数
t = quotesdf[(quotesdf.index >= u'2016-02-01') & (quotesdf.index < u'2016-03-01')]
print len(t)

#统计近一年每个月的股票开盘天数
import time
listtemp = []
for i in range(0,len(quotesdf)):
    temp = time.strptime(quotesdf.index[i],"%Y-%m-%d")
    listtemp.append(temp.tm_mon)
print listtemp
tempdf = quotesdf.copy()
tempdf['month'] = listtemp
print tempdf['month'].value_counts()
#.value_counts()每个月的天数相加
```

### 5.6分组

```python
#统计近一年每个月的股票开盘天数？
print tempdf.groupby('month').count()
#天数
#print tempdf.groupby('month').count().month
#AttributeError: 'DataFrame' object has no attribute 'month'
print tempdf.groupby('month').count().open

#如果更高效统计近一年每个月的总成交量？
print tempdf.groupby('month').sum().volume

#更高效的统计近一年每个月的总成交量，对于大量数据计数效率就会比较前一个高
g = tempdf.groupby('month')
gvolume = g['volume']
print gvolume.sum()
```
